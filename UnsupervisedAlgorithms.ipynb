## Algoritmo de K-Means

- Primer prubea de K-Means con 2 componentes principales y 3, 4 y 5 clusteres

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score, normalized_mutual_info_score, calinski_harabasz_score


# Preprocesamiento: Seleccionar características numéricas, escalar y aplicar PCA
features = df.select_dtypes(include=[np.number])
scaler = StandardScaler()
scaled_features = scaler.fit_transform(features)

# Aplicación de PCA para reducir a 2 componentes principales
pca = PCA(n_components=2)
pca_features = pca.fit_transform(scaled_features)

# Evaluación de K-means con diferentes números de clusters
for n_clusters in [3, 4, 5]:
    kmeans = KMeans(n_clusters=n_clusters, random_state=42)
    kmeans_labels = kmeans.fit_predict(pca_features)

    silhouette_avg = silhouette_score(pca_features, kmeans_labels)
    inertia = kmeans.inertia_
    calinski_harabasz = calinski_harabasz_score(pca_features, kmeans_labels)

    print(f"K-means con {n_clusters} clusters")
    print(f"Índice de Silhouette: {silhouette_avg:.2f}")
    print(f"Inercia: {inertia:.2f}")
    print(f"Calinski-Harabasz Score: {calinski_harabasz:.2f}")
    print("-"*30)

    plt.figure(figsize=(8, 6))
    plt.scatter(pca_features[:, 0], pca_features[:, 1], c=kmeans_labels, cmap='viridis', marker='o')
    plt.title(f'K-means con {n_clusters} clusters\nSilhouette Score: {silhouette_avg:.2f}')
    plt.xlabel('PCA Component 1')
    plt.ylabel('PCA Component 2')
    plt.show()
```

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
import plotly.express as px


# Preprocesamiento: Seleccionar características numéricas, escalar y aplicar PCA
features = df.select_dtypes(include=[np.number])
scaler = StandardScaler()
scaled_features = scaler.fit_transform(features)

# Aplicación de PCA para reducir a 3 componentes principales
pca = PCA(n_components=3)
pca_features = pca.fit_transform(scaled_features)

# Aplicación de K-means con 3 clusters
kmeans = KMeans(n_clusters=3, random_state=42)
kmeans_labels = kmeans.fit_predict(pca_features)

# Crear un DataFrame para Plotly
plot_df = pd.DataFrame(data=pca_features, columns=['PCA Component 1', 'PCA Component 2', 'PCA Component 3'])
plot_df['Cluster'] = kmeans_labels

# Visualización interactiva en 3D con Plotly
fig = px.scatter_3d(plot_df, x='PCA Component 1', y='PCA Component 2', z='PCA Component 3',
                    color='Cluster', labels={'color': 'Cluster'})
fig.update_traces(marker=dict(size=5))
fig.update_layout(title='K-means con 3 clusters')
fig.show()
```

- Misma Prueba con 3, 4 y 5 clusteres

```python
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score, calinski_harabasz_score
import plotly.express as px

# Seleccionar las tres características manualmente
selected_features = df[['Current_J0', 'Temperature_T0', 'Speed_J0']]

# Función para aplicar K-means, evaluar métricas y generar gráficos interactivos
def apply_kmeans_and_plot(n_clusters, features):
    kmeans = KMeans(n_clusters=n_clusters, random_state=42)
    kmeans_labels = kmeans.fit_predict(features)

    silhouette_avg = silhouette_score(features, kmeans_labels)
    inertia = kmeans.inertia_
    calinski_harabasz = calinski_harabasz_score(features, kmeans_labels)

    print(f"K-means con {n_clusters} clusters")
    print(f"Índice de Silhouette: {silhouette_avg:.2f}")
    print(f"Inercia: {inertia:.2f}")
    print(f"Calinski-Harabasz Score: {calinski_harabasz:.2f}")
    print("-" * 30)

    # Crear un DataFrame para Plotly
    plot_df = pd.DataFrame(data=features, columns=['Current_J0', 'Temperature_T0', 'Speed_J0'])
    plot_df['Cluster'] = kmeans_labels

    # Visualización interactiva en 3D con Plotly
    fig = px.scatter_3d(plot_df, x='Current_J0', y='Temperature_T0', z='Speed_J0',
                        color='Cluster', labels={'color': 'Cluster'})
    fig.update_traces(marker=dict(size=5))
    fig.update_layout(title=f'K-means con {n_clusters} clusters')
    fig.show()

# Aplicar y graficar K-means con 3 clusters
apply_kmeans_and_plot(3, selected_features)

# Aplicar y graficar K-means con 4 clusters
apply_kmeans_and_plot(4, selected_features)

# Aplicar y graficar K-means con 5 clusters
apply_kmeans_and_plot(5, selected_features)
```

## Algoritmo DBSCAN

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.cluster import DBSCAN
import matplotlib.pyplot as plt

# Paso 1: Preprocesamiento y Reducción de Dimensionalidad
# Seleccionar características numéricas y escalar los datos
features = df.select_dtypes(include=[np.number])
scaler = StandardScaler()
scaled_features = scaler.fit_transform(features)

# Aplicar PCA para reducir a 2 componentes principales
pca = PCA(n_components=2)
pca_features = pca.fit_transform(scaled_features)

# Paso 2: Aplicación de DBSCAN
dbscan = DBSCAN(eps=0.4, min_samples=10)  # Ajusta estos valores según el comportamiento de tus datos
dbscan_labels = dbscan.fit_predict(pca_features)

# Paso 3: Visualización en 2D
# Crear un DataFrame para graficar
plot_df = pd.DataFrame(data=pca_features, columns=['PCA Component 1', 'PCA Component 2'])
plot_df['Cluster'] = dbscan_labels

# Graficar los resultados en 2D
plt.figure(figsize=(10, 7))

# Graficar los clusters detectados por DBSCAN
plt.scatter(plot_df['PCA Component 1'][plot_df['Cluster'] != -1],
            plot_df['PCA Component 2'][plot_df['Cluster'] != -1],
            c=plot_df['Cluster'][plot_df['Cluster'] != -1],
            cmap='plasma', s=50, label='Clusters')

# Graficar los puntos de ruido en negro con un marcador 'x'
plt.scatter(plot_df['PCA Component 1'][plot_df['Cluster'] == -1],
            plot_df['PCA Component 2'][plot_df['Cluster'] == -1],
            color='black', marker='x', s=50, label='Ruido')

plt.title('DBSCAN Clustering (2D PCA)')
plt.xlabel('PCA Component 1')
plt.ylabel('PCA Component 2')
plt.colorbar(label='Cluster')
plt.legend()
plt.show()
```

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.cluster import DBSCAN
import plotly.express as px

# Paso 1: Preprocesamiento y Reducción de Dimensionalidad
features = df.select_dtypes(include=[np.number])
scaler = StandardScaler()
scaled_features = scaler.fit_transform(features)

# Aplicar PCA para reducir a 3 componentes principales
pca = PCA(n_components=3)
pca_features = pca.fit_transform(scaled_features)

# Paso 2: Aplicación de DBSCAN
dbscan = DBSCAN(eps=0.7, min_samples=5)
dbscan_labels = dbscan.fit_predict(pca_features)

# Paso 3: Visualización interactiva en 3D con Plotly
plot_df = pd.DataFrame(data=pca_features, columns=['PCA Component 1', 'PCA Component 2', 'PCA Component 3'])
plot_df['Cluster'] = dbscan_labels

# Graficar los resultados en 3D con un mapa de colores continuo
fig = px.scatter_3d(plot_df,
                    x='PCA Component 1',
                    y='PCA Component 2',
                    z='PCA Component 3',
                    color='Cluster',
                    labels={'color': 'Cluster'},
                    title='DBSCAN Clustering (3D PCA)',
                    color_continuous_scale=px.colors.sequential.Plasma)

# Ajustar los marcadores para los puntos de ruido
fig.update_traces(marker=dict(size=5))

# Mostrar el gráfico interactivo
fig.show()
```

```python
from sklearn.metrics import silhouette_score, calinski_harabasz_score

# Filtrar los datos que no han sido marcados como ruido (etiquetados como -1 por DBSCAN)
non_noise_mask = dbscan_labels != -1
non_noise_features = pca_features[non_noise_mask]
non_noise_labels = dbscan_labels[non_noise_mask]

# Verificar si hay suficientes clusters (al menos 2) para calcular las métricas
if len(set(non_noise_labels)) > 1:
    # Índice de Silhouette
    silhouette_avg = silhouette_score(non_noise_features, non_noise_labels)
    print(f"Índice de Silhouette (sin ruido): {silhouette_avg}")

    # Puntuación de Calinski-Harabasz
    calinski_harabasz = calinski_harabasz_score(non_noise_features, non_noise_labels)
    print(f"Puntuación de Calinski-Harabasz (sin ruido): {calinski_harabasz}")
else:
    print("No se pueden calcular las métricas debido a la falta de clusters distintos (excluyendo ruido).")
```

## Algoritmo BIRCH

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.cluster import Birch
import plotly.express as px

# Paso 1: Preprocesamiento y Reducción de Dimensionalidad
features = df.select_dtypes(include=[np.number])
scaler = StandardScaler()
scaled_features = scaler.fit_transform(features)

# Aplicar PCA para reducir a 3 componentes principales
pca = PCA(n_components=3)
pca_features = pca.fit_transform(scaled_features)

# Paso 2: Aplicación de BIRCH
birch = Birch(n_clusters=5)
birch_labels = birch.fit_predict(pca_features)

# Paso 3: Visualización interactiva en 3D con Plotly
plot_df = pd.DataFrame(data=pca_features, columns=['PCA Component 1', 'PCA Component 2', 'PCA Component 3'])
plot_df['Cluster'] = birch_labels

# Graficar los resultados en 3D con un mapa de colores
fig = px.scatter_3d(plot_df,
                    x='PCA Component 1',
                    y='PCA Component 2',
                    z='PCA Component 3',
                    color='Cluster',
                    labels={'color': 'Cluster'},
                    title='BIRCH Clustering (3D PCA)',
                    color_continuous_scale=px.colors.sequential.Viridis)

# Ajustar los marcadores
fig.update_traces(marker=dict(size=5))

# Mostrar el gráfico interactivo
fig.show()
```

```python
from sklearn.metrics import silhouette_score, calinski_harabasz_score

# Índice de Silhouette
silhouette_avg = silhouette_score(pca_features, birch_labels)
print(f"Índice de Silhouette: {silhouette_avg}")

# Inercia (distancia dentro de los clusters)
inertia = np.sum(np.min(np.linalg.norm(pca_features[:, np.newaxis] - birch.subcluster_centers_, axis=2), axis=1))
print(f"Inercia: {inertia}")

# Puntuación de Calinski-Harabasz
calinski_harabasz = calinski_harabasz_score(pca_features, birch_labels)
print(f"Puntuación de Calinski-Harabasz: {calinski_harabasz}")
```
